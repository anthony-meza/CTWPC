```python
def remove_seasonal(ds):
    climatology = ds.groupby("time.month").mean("time")
    anomalies = ds.groupby("time.month") - climatology
    return anomalies
def remove_daily_climatology(ds):
    climatology = ds.groupby("time.dayofyear").mean("time")
    anomalies = ds.groupby("time.dayofyear") - climatology
    return anomalies
```

```python
#read in Amaya Coastal wave index 
ikw_Idx = xr.open_dataset(r"../GLORYS_processed/ikw.daily.glorys.1993-2018.nc")
ikw_Idx["PC1"] = ikw_Idx.ikw_amp * np.cos(np.deg2rad(ikw_Idx.ikw_ph))
ikw_Idx["PC2"] = ikw_Idx.ikw_amp * np.sin(np.deg2rad(ikw_Idx.ikw_ph))
```

```python
%%time 
def _preprocess(ds):
    return ds.sel(latitude = slice(-2, 60), longitude = slice(-150, -75), depth = slice(0, 1))
ds = xr.open_mfdataset(
        GLORYS_data_dir("GLORYS_SFC_Anomalies.nc"),
        data_vars="minimal",
        coords="minimal",
        compat="override",
        preprocess=_preprocess,
        parallel=True,
        chunks={"time":-1, "latitude":"auto", "longitude":"auto"},
        engine="netcdf4")
```

```python
%%time 
ds = ds[["thetao", "so", "zos"]]
ds["time"] = ds.indexes['time'].normalize()
```

```python
%%time 
ds_climatology = xr.open_mfdataset(GLORYS_data_dir("GLORYS_SFC_Climatology.nc"))
```

```python
%%time 
anomalies = ds.groupby("time.dayofyear") - ds_climatology
```

```python
%%time 
theta_full = ds.thetao.sel(depth = 0, latitude = 30, longitude = -116, method = "nearest").compute()
theta_anomaly = anomalies.thetao.sel(depth = 0, latitude = 30, longitude = -116, method = "nearest").compute()

eta_full = ds.zos.sel(latitude = 30, longitude = -116, method = "nearest").compute()
eta_anomaly = anomalies.zos.sel(latitude = 30, longitude = -116, method = "nearest").compute()
```

```python
%%time 
#check that seasonal cycle has been removed 
fig, axes = plt.subplots(ncols = 2, nrows=2, figsize = (10, 3.5), sharex = True)
axs = axes.flatten()
eta_full.plot(ax = axs[0])
eta_anomaly.plot(ax = axs[1], label = "Daily Climatology Removed")
axs[0].set_xlabel("");
axs[0].set_title("SSH @ 30N, 116W"); axs[1].set_title("SSH Anomaly @ 30N, 116W")
[ax.set_ylabel("[cm]") for ax in axs[0:2]]

theta_full.plot(ax = axs[2])
theta_anomaly.plot(ax = axs[3], label = "Daily Climatology Removed")
axs[1].set_xlabel("");
axs[2].set_title("SST @ 30N, 116W"); axs[3].set_title("SST Anomaly @ 30N, 116W")
[ax.set_ylabel("[°C]") for ax in axs[2:]]
```

```python
%%time
theta_anom = anomalies.thetao.isel(depth = 0) #SST anomaly
so_anom = anomalies.so.isel(depth = 0)  #salinity anomaly
eta_anom = anomalies.zos #SSH anomaly
```

```python
#much faster than xr.corr
def timecorr(x, y):
    x_nomean = x - x.mean(dim = "time")
    y_nomean = y - y.mean(dim = "time")
                                            
    cov_xy = (x_nomean * y_nomean).mean(dim = "time")
    std_x = x_nomean.std(dim = "time")
    std_y = y_nomean.std(dim = "time")

    return cov_xy / (std_x * std_y)
```

```python
%%time
#correlate geophysical variables with coastal wave index amplitude
corr_theta_eta = timecorr(theta_anom, eta_anom).compute()
```

```python
fig, axs = plt.subplots(ncols = 1, figsize = (5, 5))
corr_theta_eta.plot.pcolormesh(ax = axs, vmin = -1, vmax = 1, cmap = cm.balance, 
                               add_colorbar = True)
```

```python
%%time
#correlate geophysical variables with coastal wave index amplitude
corr_theta = timecorr(theta_anom, ikw_Idx.ikw_amp).compute()
corr_eta = timecorr(eta_anom, ikw_Idx.ikw_amp).compute()
corr_etasq_anom = timecorr(eta_anom**2, ikw_Idx.ikw_amp).compute()
```

```python
%%time
#correlate geophysical variables with coastal wave index amplitude
corr_theta_PC1 = timecorr(theta_anom, ikw_Idx["PC1"]).compute()
corr_eta_PC1 = timecorr(eta_anom, ikw_Idx["PC1"]).compute()
corr_etasq_anom_PC1 = timecorr(eta_anom**2, ikw_Idx["PC1"]).compute()
```

```python
%%time
#correlate geophysical variables with coastal wave index amplitude
corr_theta_PC2 = timecorr(theta_anom, ikw_Idx["PC2"]).compute()
corr_eta_PC2 = timecorr(eta_anom, ikw_Idx["PC2"]).compute()
corr_etasq_anom_PC2 = timecorr(eta_anom**2, ikw_Idx["PC2"]).compute()
```

```python
%%time 
#plot the R-squared values
vmin = -0.2
vmax = 0.2
fig, axs = plt.subplots(ncols = 3, figsize = (15, 5))
corr_theta.plot.pcolormesh(ax = axs[0], vmin = vmin, vmax = vmax, cmap = cm.balance, add_colorbar = False)
corr_eta.plot.pcolormesh(ax = axs[1], vmin = vmin, vmax = vmax, cmap = cm.balance, add_colorbar = False)
cf = corr_etasq_anom.plot.pcolormesh(ax = axs[2], vmin = vmin, vmax = vmax, cmap = cm.balance, add_colorbar = False)
fig.suptitle("Correlation between Coastal Wave Index")
fig.colorbar(cf, ax = axs)

axs[0].set_title("SST Anomaly")
axs[1].set_title("SSH Anomaly")
axs[2].set_title("Wave Activity")
```

```python
%%time 
#plot the R-squared values
vmin = -0.2
vmax = 0.2
fig, axs = plt.subplots(ncols = 3, figsize = (15, 5))
corr_theta_PC1.plot.pcolormesh(ax = axs[0], vmin = vmin, vmax = vmax, cmap = cm.balance, add_colorbar = False)
corr_eta_PC1.plot.pcolormesh(ax = axs[1], vmin = vmin, vmax = vmax, cmap = cm.balance, add_colorbar = False)
cf = corr_etasq_anom_PC1.plot.pcolormesh(ax = axs[2], vmin = vmin, vmax = vmax, cmap = cm.balance, add_colorbar = False)
fig.suptitle("Correlation between PC1 of Coastal Wave Index")
fig.colorbar(cf, ax = axs)

axs[0].set_title("SST Anomaly")
axs[1].set_title("SSH Anomaly")
axs[2].set_title("Wave Activity")
```

```python
%%time 
#plot the R-squared values
vmin = -0.2
vmax = 0.2
fig, axs = plt.subplots(ncols = 3, figsize = (15, 5))
corr_theta_PC2.plot.pcolormesh(ax = axs[0], vmin = vmin, vmax = vmax, cmap = cm.balance, add_colorbar = False)
corr_eta_PC2.plot.pcolormesh(ax = axs[1], vmin = vmin, vmax = vmax, cmap = cm.balance, add_colorbar = False)
cf = corr_etasq_anom_PC2.plot.pcolormesh(ax = axs[2], vmin = vmin, vmax = vmax, cmap = cm.balance, add_colorbar = False)
fig.suptitle("Correlation between PC2 of Coastal Wave Index")
fig.colorbar(cf, ax = axs)

axs[0].set_title("SST Anomaly")
axs[1].set_title("SSH Anomaly")
axs[2].set_title("Wave Activity")
```

```python
%%time
#correlate geophysical variables with coastal wave index amplitude
rsq_theta= (corr_theta_PC1**2) + (corr_theta_PC2**2)
rsq_eta = (corr_eta_PC1**2) + (corr_eta_PC2**2)
rsq_etasq_anom = (corr_etasq_anom_PC1**2) + (corr_etasq_anom_PC2**2)
```

```python
%%time 
#plot the R-squared values
vmin = 0.0
vmax = 0.2
fig, axs = plt.subplots(ncols = 3, figsize = (15, 5))
rsq_theta.plot.pcolormesh(ax = axs[0], vmin = vmin, vmax = vmax, cmap = cm.amp, add_colorbar = False)
rsq_eta.plot.pcolormesh(ax = axs[1], vmin = vmin, vmax = vmax, cmap = cm.amp, add_colorbar = False)
cf = rsq_etasq_anom.plot.pcolormesh(ax = axs[2], vmin = vmin, vmax = vmax, cmap = cm.amp, add_colorbar = False)
fig.suptitle("Proportion Variance Explained by Coastal Wave Index (PC1 + PC2)")
fig.colorbar(cf, ax = axs)

axs[0].set_title("SST Anomaly")
axs[1].set_title("SSH Anomaly")
axs[2].set_title("Wave Activity")
```
