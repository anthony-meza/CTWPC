```python
def rm_each_month_mean(ds):
    xy_mean = ds["thetao"].mean(dim=['latitude', 'longitude'], skipna=True)
    z_mean = xy_mean.mean(dim = "depth")
    return xy_mean - z_mean
def remove_seasonal(ds):
    climatology = ds.groupby("time.month").mean("time")
    anomalies = ds.groupby("time.month") - climatology
    return anomalies.drop("month")
def remove_daily_climatology(ds):
    climatology = ds.groupby("time.dayofyear").mean("time")
    anomalies = ds.groupby("time.dayofyear") - climatology
    return anomalies.drop("dayofyear")
def remove_season_mean(ds):
    climatology = ds.groupby("time.season").mean("time")
    anomalies = ds.groupby("time.season") - climatology
    return anomalies.drop("season")
```

```python
#areal average function
def path_distance(LAT, LON):
    #compute distance along path 
    central_node = (-140, 0)
    return haversine(central_node[0], central_node[1], LON, LAT)
    
def along_coast_average(ds, mask, LON, LAT): 
    mask[mask == 0] = np.nan
    ds_mask = ds * mask
    pts = []
    #if past equator do an average in x 
    ds_eq = ds_mask * (LAT <= 2)
    ds_eq[ds_eq == 0] = np.nan
    eq_avg = np.nanmean(ds_eq, axis = 0)
    pts.append(eq_avg)
    #need to resolve Gulf of California Eventually 
    ds_noteq = ds_mask * ((LAT >= 3) & (LAT <= 35))
    ds_noteq[ds_noteq == 0] = np.nan
    noteq_avg = np.nanmean(ds_noteq, axis = 1)
    pts.append(noteq_avg)
    
    ds_noteq = ds_mask * (LAT >= 40)
    ds_noteq[ds_noteq == 0] = np.nan
    noteq_avg = np.nanmean(ds_noteq, axis = 1)
    pts.append(noteq_avg)
    
    return np.concatenate(pts)

def ds_along_coast_average(ds, mask, LON, LAT): 
    mask[mask == 0] = np.nan
    ds_mask = ds * mask
    pts = []
    #if past equator do an average in x 
    ds_eq = ds_mask * (LAT <= 2)
    ds_eq.values[ds_eq.values == 0] = np.nan
    eq_avg = ds_eq.mean(dim = "latitude")
    pts.append(eq_avg)
    #need to resolve Gulf of California Eventually 
    ds_noteq = ds_mask * ((LAT >= 3) & (LAT <= 35))
    ds_noteq.values[ds_noteq.values == 0] = np.nan
    noteq_avg = ds_noteq.mean(dim = "longitude")
    pts.append(noteq_avg)
    
    ds_noteq = ds_mask * (LAT >= 40)
    ds_noteq.values[ds_noteq.values == 0] = np.nan
    noteq_avg = ds_noteq.mean(dim = "longitude")
    pts.append(noteq_avg)
    
    return np.concatenate(pts)
```

```python
def read_ds_ocn(fname): #very little memory footprint, 1 day = 500kib
    with xr.open_dataset(fname) as ds:
        ds_new = ds[["thetao"]].sel(depth = 25, method = "nearest")
        ds_new = ds_new.sel(latitude = slice(-2, 60), longitude = slice(-140, -75) )
        return ds_new
    
def read_GLORYS_subset(years, read_ocn):
    n_cores = 5
    dsets = []
    
    for year in years:
        print(year)
        files = np.empty(0, dtype = str)
        for month in months:
            file_path = GLORYS_dir() + "/" + year + "/" + month
            file_list = natsorted([str(result) for result in list(Path(GLORYS_dir() + "/" + year + "/" + month + "/").rglob("*.[nN][cC]"))])
            files = np.concatenate([files, file_list])
        files = natsorted([str(f) for f in files])
        if __name__ == '__main__':
            with Pool(n_cores) as p:
                datasets = p.map(read_ocn, files)
        datasets = [xr.concat(datasets, dim = "time")]
        dsets = dsets + datasets #about 50 gigs of data!! 
        gc.collect()
    
    return dsets
```
