```python
def islessthanline(lat, lon, f):
    temp = np.empty_like(lat, dtype = bool)
    nrow, ncol = lat.shape
    for j in range(ncol):
        temp[:, j] = (lat[:, j] < f(lon[:, j]))
        
    return temp

def isgreaterthanline(lat, lon, f):
    temp = np.empty_like(lat, dtype = bool)
    nrow, ncol = lat.shape
    for j in range(ncol):
        temp[:, j] = (lat[:, j] > f(lon[:, j]))
    return temp
```
