```python
import subprocess
subprocess
#commands to download monthly data 
cmds = ["python", "-m", "motuclient", "--motu", "https://my.cmems-du.eu/motu-web/Motu",
"--service-id", "GLOBAL_MULTIYEAR_PHY_001_030-TDS", "--product-id", "cmems_mod_glo_phy_my_0.083_P1M-m",
"--longitude-min", "-135", "--longitude-max", "-110", "--latitude-min", "20", "--latitude-max", "50",
"--date-min", "'1993-01-17 00:00:00'", "--date-max", "'1994-01-17 00:00:00'",
"--depth-min", "0.49402499198913574", "--depth-max", "318.1274108886719", "--variable", "so", "--variable", "thetao", "--out-dir", "'.'", "--out-name",
"'GLORYS_Data1993.nc'", "--user", "'ameza'", "--pwd", "'Ameza1998'"]

#only need to change year range and filename 


for year in range(1993, 2021):
    outname = "'GLORYS_Data_" + str(year) + ".nc'"
    year_min = "'" + str(year) + "-01-01 00:00:00'"
    year_max = "'" + str(year+1) + "-01-01 00:00:00'"
    cmds[18] = year_min 
    cmds[20] = year_max
    cmds[32] = outname
    subprocess.run(cmds)
```

```python
import xarray as xr
dss = []
for year in range(1993, 2020):
    inname = "GLORYS_Data_" + str(year) + ".nc"
    with xr.open_dataset(inname) as da:
        dss.append(da)
```

```python
ds = xr.concat(dss, dim = "time")
```

```python
ds = ds.sortby("time")
```

```python
ds.to_netcdf("GLORYS_Monthly_Data.nc", format="NETCDF4")
```
