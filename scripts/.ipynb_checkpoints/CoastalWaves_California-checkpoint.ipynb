```python
GLORYS_dir = r"/vortexfs1/share/cdx/hseo/Data/MERCATOR/NEP_daily" 
results = [str(result) for result in list(Path(GLORYS_dir).rglob("*.[nN][cC]"))] #get all files
results = natsorted(results) #sort all files 

years = natsorted(list(set([result[41:45] for result in results])))
months = natsorted(['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12'])
```

```python
#trying to recreate line 90 in California
# y = mx + b 
x0 = -117; y0 = 33.5; x1 = -119; y1 = 32.5
m = ((y0 - y1) / (x0 - x1))
b = y0- x0*m 
line = lambda x : m*x + b
xs_C = np.linspace(-118.3, -117.35)
ys_C = line(xs_C)
d = np.sqrt((xs_C[0]-x0)**2 + (ys_C[0]-y0)**2)
print(d)
```

```python
def _preprocess(ds):
    return ds[["thetao", "so"]].sel(latitude = slice(-2, 60), 
                                     longitude = slice(-150, -75))
```

```python
ds = xr.open_mfdataset(
        results,
        data_vars="minimal",
        coords="minimal",
        compat="override",
        preprocess=_preprocess,
        parallel=True,
        engine="scipy")
```

```python
# %%time Must uses axes when plotting with xarray, else there will be a memory leak 
fig, axes = plt.subplots(figsize = (10, 5))
lons = ds.longitude.values
lats = ds.latitude.values
LONS, LATS = np.meshgrid(lons, lats)

vals = ds.thetao.isel(time = 0, depth = 0).compute()
axes.plot(xs_C, ys_C, color = "black"); 
vals.plot(ax = axes)
axes.set_xlim(-122, -115); 
axes.set_ylim(31, 35); 

fig.tight_layout()
```

```python
i_ind, j_ind = get_latlon_idx(ds, xs_C, ys_C)
#sample along the line 
prev_ind = (np.nan, np.nan)
lons_list = []
lats_list = []
i_ind = []
j_ind = []
tot_dist = []
for (x, y) in zip(xs_C, ys_C): 
    new_ind = where_haversine_min(x, y, LONS, LATS)
    i_ind.append(new_ind[0])
    j_ind.append(new_ind[1])
    lons_list.append(LONS[new_ind[0], new_ind[1]])
    lats_list.append(LATS[new_ind[0], new_ind[1]])
    tot_dist.append(haversine_np(lons_list[-1], lats_list[-1], xs_C[-1], ys_C[-1]))

tot_dist, uniq_dist = np.unique(tot_dist, return_index = True)
i_ind = np.array(i_ind)[uniq_dist]
j_ind =np.array(j_ind)[uniq_dist]
lons_list = np.array(lons_list)[uniq_dist]
lats_list = np.array(lats_list)[uniq_dist]
```

```python
lat_inds = xr.DataArray(i_ind, dims="distance")
lon_inds = xr.DataArray(j_ind, dims="distance")
ds = ds.isel(longitude = lon_inds, latitude = lat_inds)
ds["distance"] = tot_dist
ds.distance.attrs["units"] = "km"
```
